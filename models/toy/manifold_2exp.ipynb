```python
import numpy as np
from matplotlib.mlab import PCA
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as p3
from mpl_toolkits.mplot3d import proj3d


def sumexps(t):
    return lambda theta1, theta2: np.exp(-theta1*t) + np.exp(-theta2*t)

def diffexps(t):
    return lambda theta1, theta2: np.exp(-theta1*t) - np.exp(-theta2*t)

ts = [0.5, 1, 2]

# f: data function
fs = [sumexps(t) for t in ts] + [diffexps(t) for t in ts]
    
# make parameter grid
theta1s, theta2s = np.logspace(-3,4,500), np.logspace(-3,4,500)
theta1ss, theta2ss = np.meshgrid(theta1s, theta2s)

# generate points for data manifold
ys = [f(theta1ss, theta2ss) for f in fs]
```

```python
def pca(ys, alpha):
    """
    alpha: the damping parameter for the exponential difference data
    """
    # doc for function PCA: https://www.clear.rice.edu/comp130/12spring/pca/pca_docs.shtml
    p = PCA(np.array([y.flatten() for y in ys[:3]] +
                     [y.flatten()*alpha for y in ys[3:]]).T, standardize=False)
    
    # z denotes PCA-transformed data
    zs = [p.Y.T[i].reshape(ys[i].shape)+ys[i].mean() for i in range(len(ys))]
    
    return zs
```

```python
def plot_manifold(ys, alpha=1):
    fig = plt.figure()

    ax = fig.gca(projection='3d')
    ax.set_aspect("equal")

    ax.plot_surface(*ys[:3], color='b', alpha=0.5, 
                    shade=False, edgecolor='none')

    ax.set_xlim(0,2)
    ax.set_ylim(0,2)
    ax.set_zlim(0,2)
    ax.set_title(r'$\alpha$=%s'%str(alpha))

    plt.show()

for alpha in [0.01, 0.1, 0.2, 0.5, 1]:
    zs = pca(ys, alpha)
    plot_manifold(zs, alpha)
```
